```python
#Changing datatype of discounted price and actual price
orders['discounted_price'] = orders['discounted_price'].str.replace("₹",'')
orders['discounted_price'] = orders['discounted_price'].str.replace(",",'')
orders['discounted_price'] = orders['discounted_price'].astype('float64')

orders['actual_price'] = orders['actual_price'].str.replace("₹",'')
orders['actual_price'] = orders['actual_price'].str.replace(",",'')
orders['actual_price'] = orders['actual_price'].astype('float64')



```

```python
#Changing datatype and values in discount percentage
orders['discount_percentage'] = orders['discount_percentage'].str.replace('%','').astype('float64')
orders['discount_percentage'] = orders['discount_percentage'] / 100
```

```python
# Plot actual price vs rating
plt.scatter(orders['actual_price'], orders['rating'])
plt.xlabel('Actual_price')
plt.ylabel('Rating')
plt.show()
```
